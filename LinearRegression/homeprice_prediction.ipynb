# Home price Prediction Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# Let's plot this in visual way
plt.scatter(x='area', y='price', data=df)
plt.show()
```

```python
# now we will train the model and predict output

model = LinearRegression()

# For train the model we need input - (2D array) and output - (1D array)
x = df[['area']] # input

# or we can use
# x = df.drop(columns='price')

y = df['price'] # which is 1D array

model.fit(x, y)
```

```python
# Now we are ready to predict the price for new home with input as area
# In predict function we pass the input as 2D array

predict = model.predict([[3500]])
predict
```

```python
model.predict([[4000]])
```

```python
# Now we one csv file area.csv with some areas so we will predict the price for it
# and will export new csv file which will be  homeprice_prediction.csv
 
data = pd.read_csv('areas.csv')

predicted_price = model.predict(data)

predicted_price
```

```python
# Now we will add price column in dataset
data['price'] = predicted_price

data.head()
```

```python
# We are ready to export it

data.to_csv('homeprice_prediction.csv')
```
