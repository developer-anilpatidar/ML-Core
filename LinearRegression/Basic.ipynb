# What is Linear Regression?

Linear Regression is a supervised machine learning algorithm used to predict a continuous output (e.g., price, score, salary) based on one or more input features.

It tries to find the best-fit straight line that describes the relationship between input X and output Y.

Equation:

y = mx + b

•	m: slope (how much y changes with x)

•	b: intercept (value of y when x = 0)

---

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Prepare data
x = np.array([[1], [2], [3]])       # Hours studied
y = np.array([[50], [65], [70]])    # Exam score
```

```python
# Step 2: Train the model
model = LinearRegression()
model.fit(x, y)
```

```python
# Step 3: Predict
model.predict([[4]])
```

---

Show Data with Graph

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, color="blue", label="Actual Point")

model = LinearRegression()
model.fit(x, y)
y_predict = model.predict(x)

plt.plot(x, y_predict, marker="*", color='red', label="Regression Point")

# Add labels and legend
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression Example')
plt.legend()

plt.show()
```
