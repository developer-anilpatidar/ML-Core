```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
len(iris.data)
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df.columns = iris.feature_names
```

```python
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], marker='+', color='green')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], marker='*', color='red')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal width (cm)'], df0['petal width (cm)'], marker='+', color='green')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], marker='*', color='red')
```

```python
df.head()
```

```python
x = df.drop(['target', 'flower_name'], axis='columns')
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
x_test[:5]
```

```python
model.predict([[5.1,3.8,1.6,0.2]])
```

```python
#at 46 - 0 means setosa it's correct

#check for at 132 - should be 2
model.predict([[6.4,	2.8,	5.6,	2.2]])
```

```python
# it's correct
```

```python
# Now lets change the parameters

model_c = SVC(C=10)
model_c.fit(x_train, y_train)

model_c.score(x_test, y_test)
```

```python
# it's same as previous

# now check with gamma
model_g = SVC(gamma=10)
model_g.fit(x_train, y_train)

model_g.score(x_test, y_test)
```

```python
# now it's decreased

# now check with linier kernel, default kernal was rbf
model_lk = SVC(kernel='linear')
model_lk.fit(x_train, y_train)

model_lk.score(x_test, y_test)
```
