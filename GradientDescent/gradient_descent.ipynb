```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    cost_history = []

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum((y - y_predicted) ** 2)
        cost_history.append(cost)

        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

    
    # Plotting the final best fit line
    plt.scatter(x, y, color='blue', label='Data Points')
    plt.plot(x, m_curr * x + b_curr, color='red', label='Best Fit Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent Result')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Optional: Plot cost vs iterations
    plt.plot(range(iterations), cost_history, label='Cost')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Cost Reduction Over Time')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

```python
gradient_descent(x, y)
```
